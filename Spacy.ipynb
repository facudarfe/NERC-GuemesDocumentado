```python
# Read Guemes documentado dataset
tokens = []
tags = []
with open(CARPETA_COLAB+'dataset/GD-1.conll', 'r', encoding="utf-8") as f:
  for line in f:
    splitted_line = line.split()
    if len(splitted_line) == 3:
      tokens.append(splitted_line[0])
      tags.append(splitted_line[2])
```

```python
#SPACY NER
from nltk.tokenize import word_tokenize
import pandas
import spacy
import nltk
nltk.download('punkt')

class CustomTokenizer:

  def __init__(self, vocab):
    self.vocab = vocab

  def __call__(self, tokens):
    spaces = [True] * len(tokens)

    return spacy.tokens.Doc(self.vocab, words=tokens, spaces=spaces)

#Cargar modelo Spacy en Español
nlp = spacy.load('es_core_news_sm')
nlp.tokenizer = CustomTokenizer(nlp.vocab)

objeto_spacy=nlp(tokens)
entities = [(ent.text, ent.label_) for ent in objeto_spacy.ents]
print(entities)
```
