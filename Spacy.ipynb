```python
# Read Guemes documentado dataset
text = ''
with open(CARPETA_COLAB+'dataset/GD-1.txt', 'r', encoding="utf-8") as f:
  for line in f:
    text += line
```

```python
#SPACY NER
from nltk.tokenize import word_tokenize
import pandas
import spacy
import nltk
nltk.download('punkt')

class CustomTokenizer:

  def __init__(self, vocab):
    self.vocab = vocab

  def __call__(self, text):
    words = word_tokenize(text)
    spaces = [True] * len(words)

    return spacy.tokens.Doc(self.vocab, words=words, spaces=spaces)

#Cargar modelo Spacy en Español
nlp = spacy.load('es_core_news_sm')
nlp.tokenizer = CustomTokenizer(nlp.vocab)

objeto_spacy=nlp(text)
objeto_spacy.ents
```
